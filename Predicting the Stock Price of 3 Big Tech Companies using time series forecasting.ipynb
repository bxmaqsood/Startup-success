```python
def remove_outliers_iqr(df, column, multiplier=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Apply the function with a stricter multiplier, e.g., 2.5
meta_df_filtered_strict = remove_outliers_iqr(series_meta, 'Close', multiplier=1)

# Plotting the boxplot after stricter outlier removal
plt.figure(figsize=(10, 7))
plt.boxplot(meta_df_filtered_strict['Close'])
plt.title('Boxplot of META Close Prices After Stricter Outlier Removal')
plt.ylabel('Close Prices')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Function to perform the ADF test and print results
def adf_test(series, title=''):
    print(f'ADF Test for {title}')
    result = adfuller(series, autolag='AIC')
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

    if result[1] > 0.05:
        print(f'The {title} series is not stationary.\n')
        return False  # Not stationary
    else:
        print(f'The {title} series is stationary.\n')
        return True  # Stationary

# Perform ADF test for each company
is_stationary_apple = adf_test(series_apple['Close'], 'Apple')
is_stationary_google = adf_test(series_google['Close'], 'Google')
is_stationary_meta = adf_test(series_meta['Close'], 'Meta')

# Apply differencing if not stationary and retest
if not is_stationary_apple:
    series_apple['Diff_Close'] = series_apple['Close'].diff().dropna()
    adf_test(series_apple['Diff_Close'][1:], 'Differenced Apple')

if not is_stationary_google:
    series_google['Diff_Close'] = series_google['Close'].diff().dropna()
    adf_test(series_google['Diff_Close'][1:], 'Differenced Google')

if not is_stationary_meta:
    series_meta['Diff_Close'] = series_meta['Close'].diff().dropna()
    adf_test(series_meta['Diff_Close'][1:], 'Differenced Meta')


```

```python
import pandas as pd
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Function to difference the time series
def difference(dataset):
    return dataset.diff().dropna()

# Function to split data based on date
def split_data_on_date(series, train_end, test_start):
    train = series[:train_end]
    test = series[test_start:]
    return train, test

# Differencing the datasets
diff_series_apple = difference(series_apple['Close']).astype('float32')
diff_series_google = difference(series_google['Close']).astype('float32')
diff_series_meta = difference(series_meta['Close']).astype('float32')

# Splitting the data based on dates
train_end, test_start = '2022-09-30', '2022-10-01'
train_diff_apple, test_diff_apple = split_data_on_date(diff_series_apple, train_end, test_start)
train_diff_google, test_diff_google = split_data_on_date(diff_series_google, train_end, test_start)
train_diff_meta, test_diff_meta = split_data_on_date(diff_series_meta, train_end, test_start)

# Function to fit AR model and make predictions on differenced data
def fit_forecast_ar(train, test, lags):
    model = AutoReg(train, lags=lags).fit()
    predictions = model.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
    return predictions

# Number of lags for AR model
lags_apple = 2
lags_google = 2
lags_meta = 2

# Fit and forecast with AR model for each dataset
predictions_apple_ar = fit_forecast_ar(train_diff_apple, test_diff_apple, lags_apple)
predictions_google_ar = fit_forecast_ar(train_diff_google, test_diff_google, lags_google)
predictions_meta_ar = fit_forecast_ar(train_diff_meta, test_diff_meta, lags_meta)

# Calculate RMSE for each model
rmse_apple = sqrt(mean_squared_error(test_diff_apple, predictions_apple_ar))
rmse_google = sqrt(mean_squared_error(test_diff_google, predictions_google_ar))
rmse_meta = sqrt(mean_squared_error(test_diff_meta, predictions_meta_ar))

# Print RMSE for each company
print(f'Apple RMSE: {rmse_apple}')
print(f'Google RMSE: {rmse_google}')
print(f'Meta RMSE: {rmse_meta}')

# Plotting the results
plt.figure(figsize=(14, 7))

# Apple Plot
plt.subplot(311)
plt.plot(test_diff_apple.index, test_diff_apple.values, label='Actual Apple')
plt.plot(test_diff_apple.index, predictions_apple_ar, color='red', label='Predicted Apple')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
plt.xticks(rotation=45)
plt.title('Apple AR Model Forecast vs Actuals')
plt.legend()

# Google Plot
plt.subplot(312)
plt.plot(test_diff_google.index, test_diff_google.values, label='Actual Google')
plt.plot(test_diff_google.index, predictions_google_ar, color='red', label='Predicted Google')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
plt.xticks(rotation=45)
plt.title('Google AR Model Forecast vs Actuals')
plt.legend()

# Meta Plot
plt.subplot(313)
plt.plot(test_diff_meta.index, test_diff_meta.values, label='Actual Meta')
plt.plot(test_diff_meta.index, predictions_meta_ar, color='red', label='Predicted Meta')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
plt.xticks(rotation=45)
plt.title('Meta AR Model Forecast vs Actuals')
plt.legend()

# plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Function to difference the time series
def difference(dataset):
    return dataset.diff().dropna()


# Differencing the datasets
diff_series_apple = series_apple['Diff_Close'].astype('float32')
diff_series_google = series_google['Diff_Close'].astype('float32')
diff_series_meta = series_meta['Diff_Close'].astype('float32')


# Function to split data based on date
def split_data_on_date(series, train_end, test_start):
    train = series[:train_end]
    test = series[test_start:]
    return train, test

# Splitting the data based on dates
train_end, test_start = '2022-09-30', '2022-10-01'
train_diff_apple, test_diff_apple = split_data_on_date(diff_series_apple, train_end, test_start)
train_diff_google, test_diff_google = split_data_on_date(diff_series_google, train_end, test_start)
train_diff_meta, test_diff_meta = split_data_on_date(diff_series_meta, train_end, test_start)

# Number of windows for MA model
windows_apple = 37
windows_google = 73
windows_meta = 70


# Function to fit MA model and make predictions on differenced data
def fit_forecast_ma(train, test, order):
    model = ARIMA(train, order=(0, 0, order)).fit()
    predictions = model.predict(start=len(train), end=len(train)+len(test)-1, typ='levels')
    return predictions

# Fit and forecast with MA model for each dataset
predictions_apple = fit_forecast_ma(train_diff_apple, test_diff_apple, windows_apple)
predictions_google = fit_forecast_ma(train_diff_google, test_diff_google, windows_google)
predictions_meta = fit_forecast_ma(train_diff_meta, test_diff_meta, windows_meta)

# Calculate RMSE for each model
rmse_apple = sqrt(mean_squared_error(test_diff_apple, predictions_apple))
rmse_google = sqrt(mean_squared_error(test_diff_google, predictions_google))
rmse_meta = sqrt(mean_squared_error(test_diff_meta, predictions_meta))

# Print RMSE for each company
print(f'Apple RMSE: {rmse_apple}')
print(f'Google RMSE: {rmse_google}')
print(f'Meta RMSE: {rmse_meta}')

# Plotting the results
plt.figure(figsize=(14, 7))

# Apple Plot
plt.subplot(311)
plt.plot(test_diff_apple.index, test_diff_apple.values, label='Actual Apple')
plt.plot(test_diff_apple.index, predictions_apple, color='red', label='Predicted Apple')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
plt.xticks(rotation=45)
plt.title('Apple MA Model Forecast vs Actuals')
plt.legend()

# Google Plot
plt.subplot(312)
plt.plot(test_diff_google.index, test_diff_google.values, label='Actual Google')
plt.plot(test_diff_google.index, predictions_google, color='red', label='Predicted Google')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
plt.xticks(rotation=45)
plt.title('Google MA Model Forecast vs Actuals')
plt.legend()

# Meta Plot
plt.subplot(313)
plt.plot(test_diff_meta.index, test_diff_meta.values, label='Actual Meta')
plt.plot(test_diff_meta.index, predictions_meta, color='red', label='Predicted Meta')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
plt.xticks(rotation=45)
plt.title('Meta MA Model Forecast vs Actuals')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings


# Function to split data based on date
def split_data_on_date(series, train_end, test_start):
    train = series[:train_end]
    test = series[test_start:]
    return train, test


# Function to fit MA model and make predictions on differenced data
def fit_forecast_ma(train, test, p,d,q):
    model = ARIMA(train, order=(p, d, q)).fit()
    predictions = model.predict(start=len(train), end=len(train)+len(test)-1, typ='levels')
    return predictions


# evaluate combinations of p, d and q values for an ARIMA model
def evaluate_models( p_values, d_values, q_values):
    Close_1 = series_apple['Diff_Close'].astype('float32')
    Close_2 = series_google['Diff_Close'].astype('float32')
    Close_3 = series_meta['Diff_Close'].astype('float32')


    # Splitting the data based on dates
    train_end, test_start = '2022-09-30', '2022-10-01'
    train_diff_apple, test_diff_apple = split_data_on_date(Close_1, train_end, test_start)
    train_diff_google, test_diff_google = split_data_on_date(Close_2, train_end, test_start)
    train_diff_meta, test_diff_meta = split_data_on_date(Close_3, train_end, test_start)

    best_score_apple, best_cfg_apple = float("inf"), None
    best_score_google, best_cfg_google = float("inf"), None
    best_score_meta, best_cfg_meta = float("inf"), None

    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                try:
                    predictions_apple = fit_forecast_ma(train_diff_apple, test_diff_apple, p,d,q)
                    predictions_google = fit_forecast_ma(train_diff_google, test_diff_google, p,d,q)
                    predictions_meta = fit_forecast_ma(train_diff_meta, test_diff_meta, p,d,q)

                    rmse_apple = sqrt(mean_squared_error(test_diff_apple, predictions_apple))
                    rmse_google = sqrt(mean_squared_error(test_diff_google, predictions_google))
                    rmse_meta = sqrt(mean_squared_error(test_diff_meta, predictions_meta))

                    if rmse_apple < best_score_apple:
                        best_score_apple, best_cfg_apple = rmse_apple, order
                        best_predictions_apple = predictions_apple
                    print('ARIMA APPLE%s RMSE APPLE=%.3f' % (order,rmse_apple))

                    if rmse_google < best_score_google:
                        best_score_google, best_cfg_google = rmse_google, order
                        best_predictions_google = predictions_google
                    print('ARIMA GOOGLE%s RMSE GOOGLE=%.3f' % (order,rmse_google))

                    if rmse_meta < best_score_meta:
                        best_score_meta, best_cfg_meta = rmse_meta, order
                        best_predictions_meta = predictions_meta
                    print('ARIMA META%s RMSE META=%.3f' % (order,rmse_meta))
                    print('')
                except:
                    continue

    print('Best ARIMA GOOGLE%s RMSE GOOGLE=%.3f' % (best_cfg_google, best_score_google))
    print('Best ARIMA APPLE%s RMSE APPLE=%.3f' % (best_cfg_apple, best_score_apple))
    print('Best ARIMA META%s RMSE META=%.3f' % (best_cfg_meta, best_score_meta))

    # Plotting the results
    plt.figure(figsize=(14, 7))

    # Apple Plot
    plt.subplot(311)
    plt.plot(test_diff_apple.index, test_diff_apple.values, label='Actual Apple')
    plt.plot(test_diff_apple.index, best_predictions_apple, color='red', label='Predicted Apple')
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
    plt.xticks(rotation=45)
    plt.title('Apple ARMA Model Forecast vs Actuals')
    plt.legend()

    # Google Plot
    plt.subplot(312)
    plt.plot(test_diff_google.index, test_diff_google.values, label='Actual Google')
    plt.plot(test_diff_google.index, best_predictions_google, color='red', label='Predicted Google')
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
    plt.xticks(rotation=45)
    plt.title('Google ARMA Model Forecast vs Actuals')
    plt.legend()

    # Meta Plot
    plt.subplot(313)
    plt.plot(test_diff_meta.index, test_diff_meta.values, label='Actual Meta')
    plt.plot(test_diff_meta.index, best_predictions_meta, color='red', label='Predicted Meta')
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))
    plt.xticks(rotation=45)
    plt.title('Meta ARMA Model Forecast vs Actuals')
    plt.legend()

    plt.tight_layout()
    plt.show()

    return best_predictions_apple, best_predictions_google, best_predictions_meta
```
